# Basic example using `CoordStruct` class
The premise of the `CoordStruct` class is to contain and use information related to coordinate data. This is an essential component of the `FlowStruct` class, but is also useful in its own right. Below we explore the following capabilities
1. `CoordStruct` initialisation
2. Plotting
3. Computing Derivatives

```python
# Initialisation
import flowpy2 as fp2
import numpy as np


coord_dict = {'x':np.arange(100),
              'y':np.arange(200),
              'z':np.arange(150)}

cs = fp2.CoordStruct('Cartesian',
                     coord_dict)

coord_list = [np.arange(100),
              np.arange(50),
              np.arange(150)]
coord_index = ['x', 'y', 'z']

cs1 = fp2.CoordStruct('Cartesian',
                      coord_list,
                      index=coord_index)
```

```python
index = cs.coord_index('x', slice(10,80))
array = np.random.randn(100)
array[index].shape
```

```python
# plotting contour plots

#Array in xz direction to be plotted
array = np.random.randn(100,150)

ax = cs.pcolormesh('xz', array, cmap='bwr')

ax = cs.pcolormesh('zx', array.T, cmap='bwr')

```

```python
# Line plots

# access x coordinate
x = cs['x']
array = np.sin(np.pi*x/x.max())

ax = cs.plot_line('x', array, c='b', marker='')
```

```python
# Computing derivatives
x = cs['x']
array = np.sin(np.pi*x/x.max())

d_array = cs.first_derivative('x',array)
actual = np.pi*np.cos(np.pi*x/x.max())/x.max()

ax = cs.plot_line('x', d_array, c='b', marker='')
cs.plot_line('x', d_array, c='r', ls='--', marker='', ax=ax)
```
